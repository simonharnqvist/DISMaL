```python
def expect(k, theta):
    return (1/(theta+1)) * (theta/(theta+1))**k
```

```python
state, theta0, theta1, theta2, theta1_prime, theta2_prime, t1, v, m1_star, m2_star, m1_prime_star, m2_prime_star = [1, 5, 5, 5, 5, 5, 5, 5, 0, 0, 0, 0]
s_vals = [i/10 for i in range(0,400,10)]
p = [likelihood._pdf_s(s, state, theta0, theta1, theta2, theta1_prime, theta2_prime, t1, v, m1_star, m2_star, m1_prime_star, m2_prime_star) for s in s_vals]
expect_list = [expect(k=s, theta=theta1) for s in s_vals]
state1 = pd.DataFrame({"s":s_vals, "pdf":p, "expectation":expect_list}).set_index("s")
sns.lineplot(data=state1)
```
